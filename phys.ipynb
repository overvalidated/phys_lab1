```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import norm
```

```python
try:
    measures = np.load('measures.npy')
except FileNotFoundError:
    measures = np.random.normal(7, 0.1, size=(50,))
    measures = np.round(measures*100)/100
    np.save('measures.npy', measures)
```

```python
measures.std()
```

```python
text = ""
for i in range(50):
    text += "{} & {} & {} & {} \\\\\n\\hline\n".format(i+1, measures[i], np.around(measures[i]-6.9794, 2), np.around(np.power(measures[i]-6.9794, 2), 4))
print(text)
    
```

```python
norm(measures.mean(), measures.std()).pdf(measures.mean())
```

```python
sns.set_theme()
fig = sns.histplot(measures, stat='density').get_figure()
fig.savefig('density.png', dpi=400)
```
